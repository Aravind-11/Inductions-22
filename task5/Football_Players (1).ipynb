```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Football Players.xlsx')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Club'].unique() #lets see all the clubs they play for
```

```python
for x in range(len(df['Club'])):
    if df['Club'][x] == 'Olympique Lyonnais':
        df['Club'][x] = 'OL Lyon'
    else: 
        pass
    

df['Club'].unique() #taking an example of shortening a particular team's name
```

```python
df_top_players = df[['Name','Club','Overall']]
df_top_players[:20] #top 20 players
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Club', 'Acceleration'])
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Dribbling', 'Short passing','Ball control']) #cleaning data
```

```python
df
```

```python
filtered_df = df[(df['Club'].isin(['FC Barcelona', 'Paris Saint-Germain',
       'Manchester United', 'Manchester City', 'Real Madrid','Arsenal','FC Bayern Munich'])) & 
                      (df['Nationality'].isin(['England', 'Brazil', 'Argentina',
       'Brazil', 'Italy','Spain','Germany']))]
```

```python
filtered_df #Top Players in the world
```

```python
top_df=filtered_df[filtered_df.Overall>=80.0]
```

```python
top_df
```

```python

df_plot = filtered_df.groupby(['Club', 'Nationality']).size().reset_index().pivot(columns='Club', index='Nationality', values=0).reset_index()

g = df_plot.set_index('Nationality').T.plot(kind='bar', stacked=True, 
            color=sns.color_palette())

g.figure.set_size_inches(24,10) # of nationalities in top teams
```

```python
x=top_df[['Value ($)','Wage ($)','Overall','Age']]
y=top_df['Potential']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=1
)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
x_test
```

```python
y_test
```

```python
model.score(x_test, y_test) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

```python

y_pred = knn.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
